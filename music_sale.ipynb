```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
music_sale=pd.read_csv("MusicData.csv",index_col=['index'])
music_sale.head()
```

```python
music_sale.describe()
```

```python
df_year_value = music_sale.groupby(['Year'])['Value (Actual)'].sum().reset_index()
df_year_value
```

```python
plt.figure(figsize=(12,6))
plt.plot(df_year_value['Year'],df_year_value['Value (Actual)'], marker='o', color ='g')
plt.title ('Total sale during 1973-2019', fontsize=14, fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Value')
plt.xticks(ticks=range(1973,2019,8))
plt.show()
```

```python
counts=music_sale.groupby(['Format']).count().sort_values(by=['Number of Records'],ascending=True).reset_index()
plt.figure().set_figheight(10)
plt.barh(counts['Format'],counts['Number of Records'])
plt.title('Bar chart of number File format')
plt.show()
```

```python

metric = music_sale['Metric'].value_counts()
plt.pie(metric, labels=metric.index, autopct='%1.1f%%', colors=['olivedrab','saddlebrown','rosybrown'], textprops={'color':"w"}
)
plt.axis('equal')
```
